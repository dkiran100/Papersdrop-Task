```python
def N_grams(text,n):
    tokens = re.split("\\s+",text)
    ngrams = []
    for i in range(len(tokens)-n+1):
        temps = [tokens[j] for j in range(i,i+n)]
        ngrams.append(" ".join(temps))
    return ngrams
```

```python
l1 = N_grams(txt,1)
l1
```

```python
word_count1 = {}
 
for word in l1:
    if word in word_count1:
        word_count1[word] += 1
    else:
        word_count1[word] = 1
```

```python
from collections import Counter
c = Counter(word_count1)
 
    # returns the most occurring elements
top1 = c.most_common(1)
density1 = (top1[0][1]/len(l1))*100
print(top1)
top1.append(density1)
```
